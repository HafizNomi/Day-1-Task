```python
import cv2 as cv
import numpy as np
path='img.jpg'
image=cv.imread(path)
rows,cols,channels= image.shape
for i in range(10):
    M = cv.getRotationMatrix2D((cols/2,rows/2),i,1)
    rotate = cv.warpAffine(image,M,(cols,rows))
    
    a= 1
    b = 80
    adjusted = cv.convertScaleAbs(rotate, alpha=a, beta=b)
    noisy_image = cv.GaussianBlur(rotate, (7, 7), 0)
    output_a=  f'rotate_brighted_noised{i}.png'
    output=  f'rotate_brighted{i}.png'
    output_filename = f'rotate_{i}.png'
    cv.imwrite(output,adjusted)
    cv.imwrite(output_filename,rotate)
    cv.imwrite(output_a,noisy_image)
```

```python
import cv2
import numpy as np

image = cv2.imread('vw.jpg')
x, y, width, height = 450, 360, 860, 690
color = (0, 255, 0) 
thickness = 2
image_with_rectangle = cv2.rectangle(image.copy(), (x, y), (width,height), color, thickness)
cv2.imshow('Original Image', image)
cv2.imshow('Image with Rectangle', image_with_rectangle)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
