### Question 6

Write a NumPy program to create a random vector of size 10 and sort it.
<b>Note:<\b> Do not use builtin sorting function

```python
import numpy as np
arr1=np.random.randint(15,size=(10))
print(arr1)
for i in range(0,len(arr1)):
    for j in range(0,len(arr1)):
        if arr1[i]<arr1[j]:
            temp=arr1[j]
            arr1[j]=arr1[i]
            arr1[i]=temp
print(arr1)
```
