```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
import os
import cv2 as cv
import pickle
DIR = r'C:\Users\Administrator\Desktop\MY_data\train'
CATEGORY = ['Apple', 'avocado', 'Banana', 'cherry', 'kiwi', 'mango', 'orange', 'pinenapple', 'strawberries', 'watermelon']
img_size = 32
data = []
for category in CATEGORY:
    folder = os.path.join(DIR, category)
    labels = CATEGORY.index(category)
    for img in os.listdir(folder):
        img_path = os.path.join(folder, img)
        img_arr = cv.imread(img_path)
        if img_arr is not None:
            img_arr = cv.resize(img_arr, (img_size, img_size))
            data.append([img_arr, labels])
X = []
Y = []
for feature, label in data:
    X.append(feature)
    Y.append(label)
X = np.array(X, dtype=np.float32) / 255.0
Y = np.array(Y, dtype=np.int)
pickle.dump(X, open('X.pkl', 'wb'))
pickle.dump(Y, open('y.pkl', 'wb'))
x = pickle.load(open('X.pkl', 'rb'))
y = pickle.load(open('y.pkl', 'rb'))
print(X.shape)
def modelCNN():
    model = keras.Sequential([
        keras.layers.Input(shape=(32, 32, 3)),
        keras.layers.Conv2D(64, (3, 3), activation="relu"),
        keras.layers.MaxPooling2D(2, 2),
        keras.layers.Conv2D(64, (3, 3), activation="relu"),
        keras.layers.MaxPooling2D(2, 2),
        keras.layers.Flatten(),
        keras.layers.Dense(128, activation="relu"),
        keras.layers.Dense(10, activation="softmax")  # Changed the number of units to match the number of classes
    ], name="Fruit_Classification_Model")
    model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
    return model
model = modelCNN()
model.summary()
# Make sure to use X and Y for training, not x and y
model.fit(X, Y, epochs=50, shuffle=True)








```
