# Queue

```python
import numpy as np
queue = np.zeros(10)
index = 0
for i in range(50):
    options = input("Enter Option 1.pop, 2.push, 3.exit: ")
    if options == "push":
        if index < 10:
            number = int(input("Enter number that you want to add in queue: "))
            print(number)
            queue[index] = number
            print(queue)
            index += 1
            print("Index is -----", index)
        else:
            index >= 10
            print("queue is full")
    if options == "pop":
        if index == 0:
            print("queue is empty")
        else:
            queue[0] = 0
            for i in range(1,len(queue)):
                queue[i-1]=queue[i]
            print(queue)
            index -= 1
    if options == "exit":
        break
```
